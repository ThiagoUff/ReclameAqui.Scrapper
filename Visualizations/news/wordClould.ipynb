```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stopwords_list = stopwords.words('portuguese')


```

```python
# Lendo os dados do arquivo XLSX
file_path = '../../scrapper-dataset/excel/all-companies.xlsx'

data = pd.read_excel(file_path)
```

```python
data.head()
```

```python
textos_reclamacoes = ' '.join(data['reclamacao'])
textos_sem_stopwords = ' '.join([word for word in textos_reclamacoes.split() if word.lower() not in stopwords_list])
```

```python
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(textos_sem_stopwords)

# Plote a nuvem de palavras
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
